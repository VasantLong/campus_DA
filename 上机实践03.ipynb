```python
from pyecharts_exporter import display_chart, get_chart_urls
from pyecharts.globals import CurrentConfig, ThemeType
from pyecharts import options as opts
import pandas as pd
```

## 1

```python
from pyecharts.charts import Funnel

#导入数据
df= pd.read_excel('data/货物运输量.xls',header=0)
df.head()
```

```python
print("列名：", df.columns.tolist())
```

```python
# 提取2019年的数据（不取总数）
year_2019 = df['2019年'][1:]
#year_2019 = df.iloc[:, 2] 
indicators = df.iloc[1:, 0].tolist()

# 创建2019年数据字典
data_2019 = {}
for i, indicator in enumerate(indicators):
    value = year_2019.iloc[i]
    # 过滤掉非数值数据（如NaN、注释等）
    if pd.notna(value) and isinstance(value, (int, float)):
        # 排除"注："开头的注释行和空指标
        if isinstance(indicator, str) and not indicator.startswith('注：') and indicator.strip():
            data_2019[indicator.strip()] = int(value) if not pd.isna(value) else 0

# 过滤掉值为0的数据
data_2019 = {k: v for k, v in data_2019.items() if v > 0}

print("\n2019年货运量数据：")
for k, v in data_2019.items():
    print(f"{k}: {v} 万吨")

# 准备漏斗图数据（按数值降序排列）
funnel_data = sorted(data_2019.items(), key=lambda x: x[1], reverse=True)

# 创建漏斗图
funnel = (
    Funnel(init_opts=opts.InitOpts(
        width="900px",  # 图表宽900px
        height="600px",  # 图表高600px
        theme=ThemeType.LIGHT
    ))
    .add(
        series_name="",  # 无系列名称
        data_pair=funnel_data,
        label_opts=opts.LabelOpts(
            position="outside",  # 标签位置为outside
            formatter="{b}: {c}万吨"  # 标签格式：数据名:数据值
        ),
        tooltip_opts=opts.TooltipOpts(
            trigger="item",  # 触发类型为数据项图形触发
            formatter="{b}<br/>{c}",  # 提示框格式
            is_show=True
        ),
        itemstyle_opts=opts.ItemStyleOpts(
            border_width=1,  # 边缘线宽度1px
            border_color="#fff"  # 边缘线颜色白色
        ),
        max_=max(data_2019.values())  # 设置漏斗图的最大值为数据最大值的100%
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="2019年货运量",
            subtitle="单位：万吨",
            pos_left="center",  # 标题居中
            title_textstyle_opts=opts.TextStyleOpts(font_size=16),
            subtitle_textstyle_opts=opts.TextStyleOpts(font_size=11, color="gray")
        ),
        legend_opts=opts.LegendOpts(
            orient="vertical",  # 垂直布局
            pos_left="110px",  # 距离左边110px
            pos_top="130px",  # 距离顶部130px
            is_show=True
        ),
            graphic_opts=[
        opts.GraphicGroup(
            graphic_item=opts.GraphicItem(
                left="center",
                bottom="10px",
                z=100
            ),
            children=[
                opts.GraphicText(
                    graphic_item=opts.GraphicItem(
                        left="center",
                        bottom="20px"
                    ),
                    graphic_textstyle_opts=opts.GraphicTextStyleOpts(
                        text="数据来源：国家统计局",
                        font_size=12,
                        text_align="center"
                    )
                )
            ]
        )
    ]
    )
)


display_chart(funnel)
```

## 2

```python
df= pd.read_excel('data/居民人均收支.xls',header=0)
df
```

```python
from pyecharts.charts import Radar
import math


# 提取2016-2020年的消费支出数据（使用位置索引）
# 数据位置：第6行居民消费支出，第8行城镇消费支出，第10行农村消费支出
# 年份列：2016在第6列（索引5），2020在第2列（索引1）
resident_data = df.iloc[6, [5, 4, 3, 2, 1]].values.tolist()  # 2016-2020
urban_data = df.iloc[8, [5, 4, 3, 2, 1]].values.tolist()
rural_data = df.iloc[10, [5, 4, 3, 2, 1]].values.tolist()

# 反转顺序使年份从小到大
resident_data.reverse()
urban_data.reverse()
rural_data.reverse()

print("提取的数据（2016-2020）：")
print(f"居民人均消费支出：{resident_data}")
print(f"城镇居民人均消费支出：{urban_data}")
print(f"农村居民人均消费支出：{rural_data}")

# 设置雷达图维度
years = ['2016年', '2017年', '2018年', '2019年', '2020年']
max_value = math.ceil(max(resident_data + urban_data + rural_data) / 5000) * 5000

# 创建雷达图
radar = (
    Radar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
    .add_schema(
        schema=[{"name": year, "max": max_value} for year in years],
        shape="circle",  # 圆形雷达图
        center=["50%", "50%"],  # 中心位置
        radius="65%"  # 半径
    )
    .add(
        series_name="居民人均消费支出",
        data=[resident_data],
        linestyle_opts=opts.LineStyleOpts(width=2, color="#5470c6"),
        areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#5470c6")
    )
    .add(
        series_name="城镇居民人均消费支出",
        data=[urban_data],
        linestyle_opts=opts.LineStyleOpts(width=2, color="#fac858"),
        areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#fac858")
    )
    .add(
        series_name="农村居民人均消费支出",
        data=[rural_data],
        linestyle_opts=opts.LineStyleOpts(width=2, color="#ee6666"),
        areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#ee6666")
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="2016—2020年居民人均消费支出情况",
            subtitle="单位：元",
            pos_left="left"  # 标题位于左边
        ),
        legend_opts=opts.LegendOpts(
            orient="vertical",  # 垂直布局
            pos_right="120px",  # 距离右边120px
            pos_top="20px",  # 距离顶部20px
        ),
        graphic_opts=[
            opts.GraphicGroup(
                graphic_item=opts.GraphicItem(
                    left="center",
                    bottom="10px",
                    z=100
                ),
                children=[
                    opts.GraphicText(
                        graphic_item=opts.GraphicItem(
                            left="center",
                            bottom="20px"
                        ),
                        graphic_textstyle_opts=opts.GraphicTextStyleOpts(
                            text="数据来源：国家统计局",
                            font_size=12,
                            text_align="center"
                        )
                    )
                ]
            )
        ]
))

display_chart(radar)
```
