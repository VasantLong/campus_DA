```python
import numpy as np
```

```python
import chardet

file = './data/分省年度数据.csv'
with open(file, 'rb') as f:
    result = chardet.detect(f.read())
    print(result)

data = np.loadtxt(file, dtype=object, delimiter=',', encoding=result['encoding'])
data
```

```python
# 提取列名和数据
headers = data[0, :]  # 年份
rows = data[1:, :]    # 数据
```

```python
print("（1）前五行数据：")
print(rows[:5])
```

```python
print(f'(2) 数据的ndim: {rows.ndim}, shape: {rows.shape}')
```

```python
gdp_data = rows[:, 1:].astype(float)
gdp_sums = np.sum(gdp_data, axis=1)  # 对每一行求和
print("(3) 每个省份的 2011-2019 年 GDP 值总和：")
for i, province in enumerate(rows[:, 0]):  # 遍历省份名称
    print(f"{province}: {gdp_sums[i]}")
```

```python
gdp_2019 = gdp_data[:, -1]  # 2019 年的 GDP 数据
total_gdp_2019 = np.sum(gdp_2019)
print("(4) 2019 年全国各省份 GDP 总值:", total_gdp_2019)
```

```python
gdp_min = np.min(gdp_data, axis=1)  # 对每一行求最小值
print("(5) 各省份 2011-2019 年 GDP 最小值：")
for i, province in enumerate(rows[:, 0]):  # 遍历省份名称
    print(f"{province}: {gdp_min[i]}")
```

```python
gdp_2018 = gdp_data[:, 1]  # 2018 年的 GDP 数据
max_gdp_2018 = np.max(gdp_2018)  # 最大值
max_province = rows[np.argmax(gdp_2018), 0]  # 最大值对应的省份名称
print(f"(6) 2018 年 GDP 值最大的省: {max_province}, GDP 值: {max_gdp_2018}")
```
