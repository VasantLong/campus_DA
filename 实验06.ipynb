## 1. 红酒品质数据预处理与可视化
要求：利用wine = pd.read_csv('c:/dataset/winequality-red.csv',sep=';')读出数据保存到一个 DataFrame对象wine中，然后进行如下操作：
* （1）用wine.info()查看数据的基本情况，并检查是否有缺失值。
* （2）检查是否有重复记录。
* （3）显示有多少条重复记录。

* （5）进行describe()统计量分析，看数据的基本属性。

```python
import pandas as pd
import matplotlib.pyplot as plt
import chardet
import seaborn as sns
```

```python
# 加载中文字体，避免中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
```

```python
# 读取数据
file1 = './data/winequality-red.csv'
with open(file1, 'rb') as f:
    result = chardet.detect(f.read())
    print(result)

wine = pd.read_csv(file1, sep=';', encoding=result['encoding'])
```

```python
print('1.(1)')
wine.info()
wine.isnull().sum()
```

```python
print('1.(2)')
wine.duplicated()
```

```python
print('1.(3)')
wine.duplicated().sum()
```

```python
print('1.(4)')
wine=wine.drop_duplicates()
print(wine)
```

```python
print('1.(5)')
wine.describe()
```

```python
print('1.(6)')
wine.boxplot(rot=36, figsize=(10,10))
```

```python
print('1.(7)')
wine.hist(figsize=(10,10),color='red')
plt.show()
```

2. 小费数据预处理与可视化
要求：利用fdata= pd.read_ excel('C:/dataset/tips_mod.xls')读出数据保存到一个 DataFrame对象fdata中，然后进行如下操作：
* （1）输出小费数据fdata并查看数据的基本情况，并检查是否有缺失值。
* （2）查看缺失值的数量
* （3）删除一行内有两个缺失值的数据
* （4）删除性别或者聚餐时间为空的行
* （5）进行describe()统计量分析，看数据的基本属性。。
* （6）绘制箱形图实现各数据分布情况的可视化。
* （7）用直方图显示数据分布特征，图形要求figsize=(10,10), color='red'。
* （8）分析小费金额和消费总额的关系。（提示：用散点图）


```python
# 读取数据
file2 = './data/tips_mod.xls'
fdata = pd.read_excel(file2)
```

```python
print('2.(1)')
fdata.info()
fdata.isnull()
```

```python
print('2.(2)')
fdata.isnull().sum()
```

```python
print('2.(3) 删除前形状：', fdata.shape)
fdata = fdata.dropna(thresh=len(fdata.columns)-1) #总列数-最大允许缺失值数量
print("删除后形状：", fdata.shape)
```

```python
print('2.(4) 删除前形状：', fdata.shape)
fdata = fdata.dropna(subset=['性别', '聚餐时间段'])
print("删除后形状：", fdata.shape)
```

```python
print('2.(5)')
fdata.describe()
```

```python
print('2.(6)')
plt.figure(figsize=(12, 6))
fdata[['消费总额', '小费', '人数']].boxplot()
plt.title('数据分布箱形图')
plt.ylabel('数值')
plt.show()
```

```python
print('2.(7)')
plt.figure(figsize=(10, 10))
fdata[['消费总额', '小费', '人数']].hist(figsize=(10,10),color='red')
plt.suptitle('数据分布直方图')
plt.tight_layout()
plt.show()
```

```python
print('2.(8)')
plt.figure(figsize=(8, 6))
plt.scatter(fdata['消费总额'], fdata['小费'], alpha=0.6)
plt.title('小费金额与消费总额关系散点图')
plt.xlabel('消费总额')
plt.ylabel('小费金额')
plt.grid(True)
plt.show()
```
