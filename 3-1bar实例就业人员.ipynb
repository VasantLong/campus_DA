```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts import options as opts
from pyecharts.charts import Bar
import pandas as pd

#导入数据
df = pd.read_csv('c:/dataset/Employedpopulation.csv',encoding='gbk')
df
```

```python
#获取数据
df1 = df.iloc[:,:0:-1]
print(df1)
data1 = list(df1.iloc[1])
lable = [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
print(data1)
print(lable)
```

```python
#绘制图形
bar = (
    Bar(opts.InitOpts(width='700px',height='400px'))
    .add_xaxis(lable)
    .add_yaxis("城镇就业", data1,bar_width=30)
    #设置全局配置
    .set_global_opts(
        #标题配置项
        title_opts=opts.TitleOpts(
            title="2007-2016年城镇就业人员情况",
            subtitle="单位：万人",pos_left='center'),
        #添加脚注
        graphic_opts=opts.GraphicGroup(
            graphic_item=opts.GraphicItem(left='6%',bottom='6%'),
            children=[
                opts.GraphicText(graphic_textstyle_opts=
                opts.GraphicTextStyleOpts(text='数据来源：国家统计局',
                                          font_size=14)
            )]
        ),
        #图例配置项
        legend_opts=opts.LegendOpts(pos_right='10%',pos_top='5%',orient='vertical'),  
        #legend_opts=opts.LegendOpts(pos_right='center',pos_bottom='-6') , 
        #x坐标轴配置项
        xaxis_opts=opts.AxisOpts(name='年份'),
        # y坐标轴配置项
        yaxis_opts=opts.AxisOpts(name='人员',name_gap=8)
        )
)
bar.load_javascript()
```

```python
bar.render_notebook()
```
