```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts import options as opts    
from pyecharts.charts import Page, Pie
L1=['教授','副教授','讲师','助教','其他']
num  = [20,30,10,12,8]
#绘制饼图
c = (
    Pie(init_opts=opts.InitOpts(width="800px")
    .add(series_name="",data_pair=[list(z) for z in zip(L1,num)])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-职称类别比例"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)

c.load_javascript()
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Page, Pie
wd = ['教授','副教授','讲师','助教','其他']
num = [20,30,10,12,8]
c = (
    Pie(init_opts=opts.InitOpts(width="800px")
    .add("",[list(z) for z in zip(wd, num)],radius = ["40%", "75%"]) # 圆环的粗细和大小  
    .set_global_opts( 
        title_opts=opts.TitleOpts(title="Pie-Radius"),
        legend_opts=opts.LegendOpts( orient="vertical", pos_top="5%", pos_left="2%" )
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)
c.load_javascript()
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Page, Pie
wd = ['教授','副教授','讲师','助教','其他']
num = [20,30,10,12,8]
c =(
    Pie(init_opts=opts.InitOpts(width="800px"))
    .add("",[list(z) for z in zip(wd, num)],radius=["20%", "55%"],center=[240,220],rosetype='radius')
    .add("",[list(z) for z in zip(wd, num)],radius=["20%", "55%"],center=[620,220],rosetype='area')
    .set_global_opts(title_opts=opts.TitleOpts(title="玫瑰图"))
)
c.load_javascript()
c.render_notebook()
```
