#### 1. 多数据系列柱状图―我国就业人员情况可视化 <br>

（2）柱形图设置图表主标题为“2007—2016年全部、城镇和乡村就业人员情况”，副标题为“单位：万人”，居中显示。   <br>

（4）设置脚注为“数据来源：国家统计局” ，位于底部靠左。<br>
