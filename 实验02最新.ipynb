```python
import pandas as pd
import numpy as np
import chardet
```

```python
print('1.')
data1 = np.array([('Mayue',3000),
                 ('Lilin',4500),
                 ('Wuyun',8000)])
adf = pd.DataFrame(data1,index = range(1,4),columns=['name','pay'])
adf['pay'] = pd.to_numeric(adf['pay'],errors='coerce')
adf
```

```python
# (1) 添加tax列和行索引为5的行
print('1.(1)')
adf['tax'] = [0.05, 0.05, 0.1]
adf.loc[5] = ['Liuxi', 5000, 0.05]
adf
```

```python
# (2) 删除行索引为5的行和tax列
print('1.(2)')
adf_del = adf.drop(5).drop('tax', axis=1)
adf
```

```python
# (3) 修改tax列和Liuxi的值
print('1.(3)')
adf['tax'] = 0.03
adf.loc[adf['name'] == 'Liuxi', ['pay', 'tax']] = [9800, 0.05]
adf
```

```python
# (4) 求工资均值
print('1.(4)')
adf['pay'] = pd.to_numeric(adf['pay'])
mean_pay = adf['pay'].mean()
print(f"平均工资: {mean_pay}")
```

```python
# (1) 显示前五行数据
file = './data/分省年度数据.csv'
with open(file, 'rb') as f:
    result = chardet.detect(f.read())
    print(result,'\n')

print('2.(1)前五行数据:')
df = pd.read_csv(file, encoding=result['encoding'])
df.head(5)
```

```python
# (2) 列出数据ndim和shape
print(f"2.(2) ndim: {df.ndim}, shape: {df.shape}")
```

```python
# (3) 每个省份2011-2019年GDP总和
df['总和'] = df.iloc[:, 1:].sum(axis=1)
print("2.(3) 各省份2011-2019年GDP总和:")
print(df[['地区', '总和']])
```

```python
# (4) 2019年全国各省份GDP总值
gdp_2019_total = df['2019年'].sum()
print(f"2.(4）2019年全国各省份GDP总值: {gdp_2019_total}")
```

```python
# (5) 各省份2011-2019年GDP最小值
df['最小值'] = df.iloc[:, 1:-1].min(axis=1)
print("2.(5）各省份2011-2019年GDP最小值:")
print(df[['地区', '最小值']])
```

```python
# (6) 2018年GDP值最大的省
max_2018_province = df.loc[df['2018年'].idxmax(), '地区']
max_2018_value = df['2018年'].max()
print(f"2.(6）2018年GDP值最大的省: {max_2018_province} {max_2018_value}")
```

```python
# (7) 湖南省2011-2019年的GDP均值
hunan_mean = df[df['地区'] == '湖南省'].iloc[:, 1:-2].mean(axis=1).values[0]
print(f"2.(7）湖南省2011-2019年GDP均值: {hunan_mean}")
```

```python
print('3.')
lending_df = pd.read_csv('./data/Lending.dat', encoding='utf-8')
lending_df['date'] = pd.to_datetime(lending_df['date'])
lending_df
```

```python
# (1) 查询1084420用户在2016年的借阅情况
print('3.(1) 1084420用户在2016年的借阅情况:')
user_1084420_2016 = lending_df[
    (lending_df['uid'] == 1084420) & 
    (lending_df['date'].dt.year == 2016)
]
user_1084420_2016
```

```python
# (2) 查询2016、2018和2019年的借阅情况
print('3.(2) 2016、2018和2019年的借阅情况:')
# years_2016_2018_2019 = lending_df[lending_df['date'].str[:4].isin(['2016', '2018', '2019'])]
years_2016_2018_2019 = lending_df[
    lending_df['date'].dt.year.isin([2016, 2018, 2019])
]
years_2016_2018_2019
```

```python
# (3) 查询借阅了1103310所借图书的用户
print('3.(3) 借阅了1103310所借图书的用户:')
users_borrowed_same = lending_df[
    lending_df['bid'].isin(
        lending_df[lending_df['uid'] == 1103310]['bid']
    ) & 
    (lending_df['uid'] != 1103310)
]

unique_users_sorted = np.sort(users_borrowed_same['uid'].unique())
unique_users_sorted
```
