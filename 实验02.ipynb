```python
import numpy as np
```

```python
import chardet
file = './data/广播电视情况.csv'
with open(file, 'rb') as f:
    result = chardet.detect(f.read())
    print(result)
```

```python
data = np.loadtxt(file, dtype=object, delimiter=',', encoding=result['encoding'])
data
```

```python
# 提取列名和数据
headers = data[0, :]
rows = data[1:, :]
years = headers[1:]
```

```python
# 通过指标名称找到对应的行数据
def get_row_by_indicator(indicator_name):
    # 遍历指标列，找到匹配的行
    for i, row in enumerate(rows):
        if row[0] == indicator_name:
            return row[1:].astype(float)  # 返回该行的数据（去掉指标名称）
    return None  # 不匹配返回 None
```

```python
years_2015_2019 = years[5:0:-1]
print("(1) 2015-2019 年的年份数据:", years_2015_2019)

broadcast_coverage_2015_2019 = get_row_by_indicator('广播节目综合人口覆盖率(%)')[5:0:-1]
print("(2) 2015-2019 年的广播节目综合人口覆盖率(%):", broadcast_coverage_2015_2019)

tv_coverage_2015_2019 = get_row_by_indicator('电视节目综合人口覆盖率(%)')[5:0:-1]
max_tv_coverage = np.max(tv_coverage_2015_2019)
print(f'(3) 2015-2019 电视节目综合人口覆盖率(%): {tv_coverage_2015_2019},最高值: {max_tv_coverage}')

tv_series_2011_2015 = get_row_by_indicator('电视剧播出部数(万部)')[-1:-6:-1]
mean_tv_series = np.mean(tv_series_2011_2015)
print("(4) 2011-2015 年的电视剧播出部数（万部）的均值:", mean_tv_series)

imported_tv_series = get_row_by_indicator('进口电视剧播出部数(部)')
imported_tv_series_total = np.sum(imported_tv_series)
print("(5) 2011-2020 年的进口电视剧播出部数（部）的总和:", imported_tv_series_total)
```
