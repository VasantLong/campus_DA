### 例题1：判断一个学生的考试成绩是否及格：如果成绩大于或等于60分，则打印“及格”，否则输出“不及格”。

```python
score = float(input("请输入成绩："))
if score >= 60:
    print("及格")
else:
    print("不及格")
```

### 例题2：毛利率的计算
毛利率(Gross Margin Rate）是反映企业盈利能力的重要指标之一。

计算公式：毛利率=毛利润/营业收入×100%=(营业收入-营业成本)/营业收入×100%。
——————————————————————————————————————————————————————————————————————————

根据公式，将计算过程分成三个步骤，IPO描述如图所示：


——————————————————————————————————————————————————————————————————————————


```python
# 1输入
income = input('请输入营业收入: ')
cost = input('请输入营业成本: ')

# 2处理、加工（类型转换、计算）
income = int(income)
cost = int(cost)
profit = (income - cost)/income*100

# 3输出
print('毛利率=', profit, '%')
```
