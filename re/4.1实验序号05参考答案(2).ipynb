3）试着绘制以上四个图构成的子图。

```python
import pandas as pd
import matplotlib.pyplot as plt
records = pd.read_csv('c:\\dataset\\lending.dat')

records['year'] = pd.to_datetime(records['date']).dt.year
results = records[['uid', 'year']].groupby('year').count()
plt.figure(figsize=(15, 8))
#第1个子图
plt.subplot(2,2,1)
plt.plot(results['uid'])

#第2个子图
plt.subplot(2,2,2)
plt.scatter(results.index, results['uid'])

#第3个子图
plt.subplot(2,2,3)
plt.bar(results.index, results['uid'])
#强制设置所有 x 刻度和标签
plt.xticks(ticks=results.index, labels=[f'{i}' for i in results.index])

#第4个子图
records['month'] = pd.to_datetime(records['date']).dt.month
results1 = records[['uid', 'month']].groupby('month').count()

plt.subplot(2,2,4)
plt.bar(results1.index, results1['uid'])
#强制设置所有 x 刻度和标签
plt.xticks(ticks=results1.index, labels=[f'{i}' for i in results1.index])

results1
results
records
```
