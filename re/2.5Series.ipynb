```python
import pandas as pd
import numpy as np
dti = pd.date_range(start='2015-1-1', end='2015-12-31',freq='B')
s = pd.Series(np.random.rand(len(dti)),index=dti)
s.head(10)
```

```python
s.index.weekday
```

```python
s[s.index.weekday == 3].head(10)
```

```python
s[s.index.weekday == 2].sum()
```

```python
list(s.resample('QE'))
```

```python
s.resample('QE').mean()
```
