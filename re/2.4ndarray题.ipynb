1、创建一个3x3 的单位矩阵。

```python
import numpy as np
Z = np.eye(3)
print(Z)
```

```python
Z = np.identity(3,int)  #np.eye 更灵活，可以生成非方阵或偏移对角线（例如 np.eye(3, 4, k=1) 生成 3×4 矩阵，且对角线向右偏移 1 列）。
print(Z)
```

2、创建一个3x3 并且值从0 到8 的矩阵。

```python
import numpy as np
Z = np.arange(9).reshape(3,3)
print(Z)
```

3、创建一个5x5 的矩阵，其中每行的数值范围从0到4。 

```python
import numpy as np
Z = np.zeros((5,5),dtype = int)
Z += np.arange(5)
print (Z)
```

```python
Z = np.full([5,5],np.arange(5))
print (Z)
```

```python
Z = np.tile(np.arange(5),(5,1)) 
print (Z)
```

```python
Z = np.array([np.arange(5) for i in range(5)])
print (Z)
```

### 4、创建一个 10x10 的二维数组，并使得边界为1，中间全为0 。

```python
import numpy as np
x=np.ones((10,10),dtype = int)
x[1:-1,1:-1]=0
print (x)
```

```python
Z = np.full([10,10],1)
Z[1:9,1:9]=0
print (Z)
```

```python
Z = np.ones((10,10),dtype = int)
for i in range(1,9):
    for j in range(1,9):
        Z[i,j]=0
print(Z)
```

5、创建一个8*8 的“棋盘”矩阵。

```python
 import numpy as np
z=np.zeros((8,8),dtype=int)
z[1::2,::2]=1    #1、3、5、7 行 ；0 、2、4、6 列的元素置为1
z[::2,1::2]=1
print (z)
```

```python
Z = np.full([10,10],0)
Z[1::2,::2]=1
Z[::2,1::2]=1
print (Z)
```

```python
Z=np.tile([[0,1],[1,0]],(4,4))  # np.tile函数可以将原矩阵横向或纵向铺开
print (Z)
```

```python
Z = np.zeros((8,8),dtype=int)
for i in range(8):
    for j in range(8):
        if(i+j)%2==1:
            Z[i,j]=1
print (Z)
```

6、创建一个长度为30的随机向量并找到它的平均值。

```python
import numpy as np
Z = np.random.random(30)
m = Z.mean()
print(Z)
print("平均值:",m)
```

7、创建一个长度为10 并且除了第五个值为1 的空向量。

```python
import numpy as np
Z = np.zeros(10,dtype = int)
Z[4] = 1
print(Z)
```
