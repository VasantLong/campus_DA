### 实验03 根据国家统计局发布的2011-2020年我国广播电视情况数据表（广播电视情况.csv，该文件部分数据如下图所示），要求从中分别获取如下数据。
（1）获取2015-2019年的年份数据。
（2）获取2015-2019年的广播节目综合人口覆盖率(%) 。
（3）获取2015-2019 电视节目综合人口覆盖率(%)的数据 ，求覆盖率最高值。 
（4）获取2011-2015年的电视剧播出部数(万部)，求均值。
（5）获取2011-2020年的进口电视剧播出部数(部)的数据，并求总和。   


```python
import numpy as np
#导入数据
data =np.loadtxt('c:/dataset/广播电视情况.csv',dtype=object,delimiter=',')
print(data.shape)
data
```

```python
#1.获取2015-2019年的年份数据。
data1 = data[0,[6,5,4,3,2]]
#data1=data[:,6:1:-1]
#data1 = data[:,[0,6,5,4,3,2]] #多显示0列
data1
```

```python
#2.获取2015-2019年的广播节目综合人口覆盖率(%)数据。
data2 = data[1,[6,5,4,3,2]].astype(float)
print(data2)
```

```python
#3.2015-2019年的电视节目综合人口覆盖率(%)的数据,求覆盖率最高值。
data3 = data[8,[6,5,4,3,2]].astype(float)
print(data3)
print(data3.max()) 
```

```python
#4.2011-2015年的电视剧播出部数(万部)，求均值。
data4 = data[17,[10,9,8,7,6]].astype(float)
print(data4)
print(data4.mean())
```

```python
#5.2011-2020年的进口电视剧播出部数(部),并求总和。
data5 = data[18,1:].astype(int)
print(data5)
print(data5.sum())
```
