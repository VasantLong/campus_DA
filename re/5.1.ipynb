```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import scipy
```

定量数据分布分析

```python
iris = datasets.load_iris()
iris.data
```

```python
iris.target
```

```python
iris_df = pd.DataFrame(iris.data)
iris_df.columns = iris.feature_names
iris_df['target'] = iris.target #target作为新列添加到数据框中
iris_df.head()
```

```python
plt.hist(iris_df.iloc[:, 0], bins=30, color='g', label='sepal length (cm)')
```

```python
scipy.stats.normaltest(iris_df.iloc[:, 0],axis=0) #正态性检验
```
