## **习题1：djidf数据**
1. 创建一个Series数据，price作为行索引
2. 获取股票价格最高信息

```python
data = pd.read_csv('./data/djidf.csv')
series = pd.Series(data['code'].values,index = data['price'].values)
print('(1)')
series
```

## **习题2：创建Series s，将2024年前五十个星期一作为随机值的索引**
1. 前三十个星期一的值求和
2. 求每个自然月的平均数

```python
import numpy as np
```

```python
date_range = pd.date_range(start='2024-01-01', periods=50*7, freq='D')
mondays = date_range[date_range.weekday == 0][:50]
np.random.seed(42)
s = pd.Series(np.random.randn(len(mondays)), index=mondays)
s.head(6)
```

```python
sum30 = s[:30].sum()
print(f'(1)前三十个星期一的值求和：{sum30}')
```

```python
print(f'(2)每个自然月的平均数：\n{s.resample('ME').mean()}')
```

```python
# 创建一个包含一周七天的日期范围
dates = pd.date_range(start='2024-01-01', periods=7)

# 查看每个日期对应的星期几数字
for date in dates:
    print(f"{date.date()} 是星期{date.weekday()+1}")

# 或者使用day_name()查看英文名称
for date in dates:
    print(f"{date.date()} 是 {date.day_name()} (weekday={date.weekday()})")
```
