```python
import requests
from bs4 import BeautifulSoup 
import re
```

```python
url1="https://www.weather.com.cn/weather1d/101030100.shtml#search"
headers={'User-Agent':'Mozilla/5.0'}
try:
    r1=requests.get(url1,headers=headers)
    r1.raise_for_status()
    r1.encoding=r1.apparent_encoding 
except:
    print("爬取失败")

soup = BeautifulSoup(r1.text, 'html.parser') 
today = soup.find('div', class_='t')
```

```python
print('1.')
weather = today.find('p', class_='wea').text
print(f"天气：{weather}")

temp = today.find('p', class_='tem').get_text(strip=True)
#temp = re.findall(r'\d+°C', temp1)
print(f"气温：{temp}")

sunset1 = today.find('p', class_='sun sunDown').get_text(strip=True)
#sunset = re.findall(r'\d{2}:\d{2}', sunset1)
#print(f"日落 {sunset[0]}")
print(sunset1)

sunrise1 = today.find('p', class_='sun sunUp').get_text(strip=True)
#sunrise = re.findall(r'\d{2}:\d{2}', sunrise1)
print(sunrise1)
```

```python
url3 = "https://www.volleyball.world/en/vnl/2019/women/resultsandranking/round1"
headers = {'User-Agent':'Mozilla/5.0'}
try:
    r3 = requests.get(url3, headers=headers)
    r3.raise_for_status()
    r3.encoding=r3.apparent_encoding
except:
    print("爬取失败")

soup = BeautifulSoup(r3.text, 'html.parser')
info = soup.find('div', class_='table__container')
```

```python
print('3.')
print(f"RANK\tTEAM{'':<12}\tTOTAL\tWON\tLOST")
for row in info.find_all('tr'):
    columns = row.find_all('td')
    if len(columns) > 0:
        rank = re.findall(r'<td class="text--extra">(\d+)', str(columns))[0]
        team = re.findall(r'<figcaption>.*?>(.*?)<\/a>', str(columns))[0]
        total = re.findall(r'<td.*?group--start">(.*?)<\/td>', str(columns))[0]
        win = re.findall(r'text--highlight">(.*?)<\/td>', str(columns))[0]
        lost = re.findall(r'<td.*?group--end">(.*?)<\/td>', str(columns))[0]
        print(f"{rank}\t{team:<20}\t{total}\t{win}\t{lost}")
```
