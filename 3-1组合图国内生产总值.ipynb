```python
from pyecharts import options as opts
from pyecharts.charts import Timeline, Bar, Pie
import pandas as pd

#导入数据
df = pd.read_csv('data/国内生产总值.csv',encoding='gbk')
df
```

```python
#获取数据
lable = [2016,2017,2018,2019,2020]
df1 = df.iloc[1:5,[5,4,3,2,1]]
df1
```

```python
data0 = list(df1.iloc[0])  #2016-2020国内生产总值(亿元)
data1 = list(df1.iloc[1])  #2016-2020第一产业增加值(亿元)
data2 = list(df1.iloc[2])  #2016-2020第二产业增加值(亿元)
data3 = list(df1.iloc[3])  #2016-2020第三产业增加值(亿元)
data1
```

```python
rate1 = list(round(df1.iloc[1]/df1.iloc[0]*100,2)) #round(x,n)方法返回浮点数x的四舍五入值。
rate2 = list(round(df1.iloc[2]/df1.iloc[0]*100,2))
rate3 = list(round(df1.iloc[3]/df1.iloc[0]*100,2))
rate1
```

```python
#定义函数
def get_year_overlap_chart(year: int) :
    # 绘制多数据系列柱形图
    bar = (
        Bar(opts.InitOpts(width="1500px", height="580px"))
        .add_xaxis(lable)
        .add_yaxis("国内生产总值", data0)
        .add_yaxis("第一产业增加值", data1)
        .add_yaxis("第二产业增加值", data2)
        .add_yaxis("第三产业增加值", data3)
        .set_global_opts(
            title_opts=opts.TitleOpts(
            title="5年期间国内生产总值及第一产业、第二产业、第三产业增加值情况",
            subtitle="单位：亿元 ",
            pos_left='left'),
            # 添加脚注
            graphic_opts=opts.GraphicGroup(
                graphic_item=opts.GraphicItem(left='4%', bottom='3%'),
                children=[
                    opts.GraphicText(graphic_textstyle_opts=
                                     opts.GraphicTextStyleOpts(text='数据来源：国家统计局',
                                     font="14px Microsoft YaHei"))]
            ),
           # 添加图例
            legend_opts=opts.LegendOpts(selected_mode='multiple',
                                          orient='vertical',
                                          is_show=True,
                                          pos_right='5',
                                          pos_bottom='55%'))
        )
    #绘制饼图
    pie = (
        Pie()
        .add(
            series_name="占比",
            data_pair=[
                ["第一产业增加值占比", rate1[year]],
                ["第二产业增加值占比", rate2[year]],
                ["第三产业增加值占比", rate3[year]]
            ],
            center=["38%", "18%"],    #饼图中心点在画布的位置
            radius="18%",
            is_clockwise=False)
        .set_global_opts(legend_opts=opts.LegendOpts(is_show=False))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}%"))
    )   
    return bar.overlap(pie)


# 创建时间线轮播多图
timeline = Timeline(init_opts=
                    opts.InitOpts(width="900px", height="600px"))  #初始化配置

for y in range(5):
    print(y)
    timeline.add(get_year_overlap_chart(year=y), time_point=str(y))  #添加图表实例和时间点
                                                                     #这里图表实例是自定义函数
timeline.add_schema(is_auto_play=True, play_interval=2000);  #timeline.add_schema函数可设置坐标轴类型、时间轴类型等
#timeline.render("d:/html/task7-1.html")
```

```python
display_chart(timeline)
```
