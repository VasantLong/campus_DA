```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import chardet

file = './data/股票数据.csv'
with open(file, 'rb') as f:
    result = chardet.detect(f.read())
    print(result)

frame = pd.read_csv(file, encoding=result['encoding'])
```

```python
# 加载中文字体，避免中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
```

```python
frame = frame.set_index('日期')    #将日期设置为索引
frame.index = pd.to_datetime(frame.index)     #将日期转换为Datetime时间类型
#print(frame)
plt.plot(frame['收盘价'])
plt.xlabel('日期')    #设置x轴标签
plt.ylabel('收盘价')    #设置y轴标签
plot_title = '股票收盘价走势图'
plt.title(plot_title)    #设置标题
plt.show()
```
