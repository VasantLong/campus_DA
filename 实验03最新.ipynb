```python
import requests
from bs4 import BeautifulSoup 
```

```python
print('1.')
url1 = "https://item.jd.com/100010088735.html"
headers = {
  'User-Agent':'Mozilla/5.0 (Windows NT 10.0; wow64)  AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'
  }
try:
    r1 = requests.get(url1,headers=headers)
    r1.raise_for_status()
    r1.encoding=r1.apparent_encoding
    print(r1.text)
except:
    print("爬取失败")
```

```python
print('2.')
url2 = "https://book.zongheng.com/chapter/912802/59031605.html"
headers = {
  'User-Agent':'Mozilla/5.0 (Windows NT 10.0; wow64)  AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'
  }
try:
    r2 = requests.get(url2,headers=headers)
    r2.raise_for_status()
    r2.encoding=r2.apparent_encoding
    print(r2.text)

    soup = BeautifulSoup(r2.text, 'lxml') #将该网页进行lxml解析
    pattern_title = soup.find("div", class_= 'name') 
    pattern_bookinfo = soup.find("div", class_= 'tags') 
    pattern_content = soup.find("div", class_= 'reader-box') #找到div标签中class属性为reader-box的内容
    print(f'小说题目：{pattern_title.string}') 
    print(f'书籍信息：{pattern_bookinfo.text}') 
    print(f'正文：\n{pattern_content.text}') 
except:
    print("爬取失败")
```
