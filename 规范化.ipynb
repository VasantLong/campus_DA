### 最小-最大规范化
* 若将来的数字超过min和max需重新定义
* 若某个数很大则规范化后值相近且均接近0


```python
# 最小-最大规范化 (df-df.min())/(df.max()-df.min())
min_max_scaler = preprocessing.minmax_scale(df)
print(min_max_scaler)
```

### z-score规范化
* 使用最多
* 处理后数据的均值为0，标准差为1

```python
# z-score规范化 (df-df.mean())/df.std()
z_scaler = preprocessing.scale(df)
print(z_scaler)
```

### 小数定标规范化
* 移动小数点位置，移动位数,取决于属性绝对值的最大值
* 常见落在[-1, 1]之间

```python
# 小数定标规范化
float_scaler = df/10**np.ceil(np.log10(df.abs().max()))
print(float_scaler)
```

## 连续属性离散化
* 将连续的数值型数据转换为离散的分类值的过程，这在数据分析中非常有用，可以简化模型、揭示数据分布模式或满足某些算法的要求。
### 分箱（binning）：
* 等宽法（数据分布均匀）：把数据按设定的箱数等分
  * 划分：如AGE范围是0-100，则分为0-20, 20-40, 40-60, 60-80, 80-100
  * 参数说明：
    * 第一个参数：要离散化的Series
    * 第二个参数：要分的箱数(或自定义的分界点列表)
    * labels：为每个箱指定标签(这里用0-4表示)
* 等频法（数据分布不均）
  * 每个区间包含近似相同数量的观测值，以分位数为边界划分
### 聚类

```python
pd.cut(df.AGE, 5, labels = range(5))  
```

```python
pd.qcut(df.AGE, 5, labels = range(5))
```
