### **例题1:判断一个学生的考试成绩是否及格:如果成绩大于或等于60分，则打印“及格”，否则输出“不及格”**

```python
score = input("请输入成绩：")
if float(score) >= 60:
    print('及格')
else:
    print('不及格')
```

### **例题2：毛利率的计算**

```python
#I
income = int(input('请输入营业收入：'))
cost = int(input('请输入营业成本：'))

#P
profit = (income - cost) / income * 100

#O
print(f'毛利率为：{profit}%')
```
