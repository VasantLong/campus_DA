```python
from pyecharts_exporter import display_chart, get_chart_urls
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType, ThemeType
from pyecharts import options as opts
import pandas as pd
```

## 1

```python
# 读取数据
df = pd.read_csv('data/用户评论词语频次.csv')

# 准备词云图数据：[(word, freq), ...]
data = [(row['关键词'], row['词频']) for _, row in df.iterrows()]

#data
```

```python
# 创建词云图
wordcloud = (
    WordCloud()#init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
    .add(
        series_name="词频",
        data_pair=data,
        word_size_range=[20, 80],          # 单词字体大小范围
        shape="pentagon",          # 词云图轮廓为pentagon
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="电商评论数据词云图",
            title_textstyle_opts=opts.TextStyleOpts(
                font_family="serif",
                font_size=30,
                color="blue"
            )
        )
    )
    .set_series_opts(
        tooltip_opts=opts.TooltipOpts(is_show=True)
    )
)

# 设置图表宽高
wordcloud.width = "900px"
wordcloud.height = "600px"

print(get_chart_urls(wordcloud))
display_chart(wordcloud)
```

## 2

```python
from pyecharts.charts import Bar, Pie, Timeline, Grid

# 读取数据
df = pd.read_csv('data/广播电视情况.csv',encoding='gbk')
df.head()
```

```python
print("列名:", df.columns.tolist())
print("\n指标列唯一值:", df['指标'].unique())
```

```python
# 提取2016-2020年电影票房数据
labels = [2015, 2016, 2017, 2018, 2019, 2020]

# 获取各项票房数据
national_box_office = []  # 全国电影票房收入
local_box_office = []     # 国产电影票房收入
foreign_box_office = []   # 进口电影票房收入

for year in labels:
    year_str = f'{year}年'
    # 获取全国电影票房收入
    try:
        national_val = df[df['指标'] == '全国电影票房收入(亿元)'][year_str].iloc[0]
        national_box_office.append(float(national_val) if pd.notna(national_val) else 0)
    except:
        national_box_office.append(0)
        
    # 获取国产电影票房收入
    try:
        local_val = df[df['指标'] == '国产电影票房收入(亿元)'][year_str].iloc[0]
        local_box_office.append(float(local_val) if pd.notna(local_val) else 0)
    except:
        local_box_office.append(0)
        
    # 获取进口电影票房收入
    try:
        foreign_val = df[df['指标'] == '进口电影票房收入(亿元)'][year_str].iloc[0]
        foreign_box_office.append(float(foreign_val) if pd.notna(foreign_val) else 0)
    except:
        foreign_box_office.append(0)

# 计算各年份国产和进口电影票房占比
local_ratio = []
foreign_ratio = []
for i in range(len(labels)):
    total = national_box_office[i]
    if total > 0:
        local_ratio.append(round(local_box_office[i] / total * 100, 2))
        foreign_ratio.append(round(foreign_box_office[i] / total * 100, 2))
    else:
        local_ratio.append(0)
        foreign_ratio.append(0)

# 定义函数创建每年的组合图表
def get_year_overlap_chart(year: int):
    # 绘制多数据系列柱形图
    bar = (
        Bar(opts.InitOpts(width="900px", height="600px"))
        .add_xaxis(['2015', '2016', '2017', '2018', '2019'])
        .add_yaxis("全国电影票房收入", national_box_office)
        .add_yaxis("国产电影票房收入", local_box_office)
        .add_yaxis("进口电影票房收入", foreign_box_office)
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="5年期间全国电影票房收入及国产、进口票房收入情况",
                subtitle="单位：亿元 ",
                pos_left='center'),
            # 添加脚注
            graphic_opts=opts.GraphicGroup(
                graphic_item=opts.GraphicItem(left='4%', bottom='3%'),
                children=[
                    opts.GraphicText(graphic_textstyle_opts=
                                     opts.GraphicTextStyleOpts(text='数据来源：国家统计局',
                                     font="14px Microsoft YaHei"))]
            ),
           # 添加图例
            legend_opts=opts.LegendOpts(selected_mode='multiple',
                                          orient='vertical',
                                          is_show=True,
                                          pos_right='5',
                                          pos_top='10%'))
        )
    
    # 绘制饼图
    pie = (
        Pie()
        .add(
            series_name="占比",
            data_pair=[
                ["国产电影票房收入占比", local_ratio[year]],
                ["进口电影票房收入占比", foreign_ratio[year]]
            ],
            center=["38%", "18%"],    # 饼图中心点在画布的位置
            radius="18%",
            is_clockwise=False)
        .set_global_opts(legend_opts=opts.LegendOpts(is_show=False))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}%"))
    )   
    return bar.overlap(pie)

# 创建时间线轮播多图
timeline = Timeline(init_opts=
                    opts.InitOpts(width="900px", height="600px"))  # 初始化配置

for y in range(5):
    timeline.add(get_year_overlap_chart(year=y), time_point=str(labels[y]))  # 添加图表实例和时间点

timeline.add_schema(is_auto_play=True, play_interval=2000)  # timeline.add_schema函数可设置坐标轴类型、时间轴类型等
```

```python
display_chart(timeline)
```
