```python
"""
PyECharts iframe 渲染器 - 高级使用示例（含动态依赖）
"""

# ========== 基础用法 ==========
from pyecharts_jupyter_exporter import display_chart, get_chart_dependencies
from pyecharts.charts import Bar, Map, WordCloud
from pyecharts import options as opts

# 1. 基础柱状图（仅依赖 echarts）
bar = (
    Bar()
    .add_xaxis(["衬衫", "毛衣", "领带"])
    .add_yaxis("商家A", [114, 55, 27])
    .set_global_opts(title_opts=opts.TitleOpts(title="基础图表"))
)

# 查看依赖信息（调试用）
info = get_chart_dependencies(bar)
print(f"依赖: {info['dependencies']}")
print(f"URLs: {info['cdn_urls']}")

# 显示图表
display_chart(bar)

```

```python
from pyecharts_jupyter_exporter import display_chart, get_chart_urls, render_chart

from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker

c = (
    Map()
    .add("商家A", [list(z) for z in zip(Faker.country, Faker.values())], "world")
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-世界地图"),
        visualmap_opts=opts.VisualMapOpts(max_=200),
    )
)
info = get_chart_urls(c)
#print(info)
#display_chart(c)
render_chart(c)
```
