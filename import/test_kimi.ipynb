```python
"""
PyECharts iframe 渲染器 - 高级使用示例（含动态依赖）
"""

# ========== 基础用法 ==========
from pyecharts_iframe_renderer import display_chart, get_chart_dependencies
from pyecharts.charts import Bar, Map, WordCloud
from pyecharts import options as opts

# 1. 基础柱状图（仅依赖 echarts）
bar = (
    Bar()
    .add_xaxis(["衬衫", "毛衣", "领带"])
    .add_yaxis("商家A", [114, 55, 27])
    .set_global_opts(title_opts=opts.TitleOpts(title="基础图表"))
)

# 查看依赖信息（调试用）
info = get_chart_dependencies(bar)
print(f"依赖: {info['dependencies']}")
print(f"URLs: {info['cdn_urls']}")

# 显示图表
display_chart(bar)

```

```python

# ========== 地图图表（依赖 china.js） ==========
from pyecharts_iframe_renderer import display_chart, CDN_CONFIGS

# 创建地图
data = [("广东", 10430.03), ("北京", 12102.53), ("上海", 14070.12)]
map_chart = (
    Map()
    .add("GDP", data, "china")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="中国地图"),
        visualmap_opts=opts.VisualMapOpts(max_=15000)
    )
)

# 查看地图图表的依赖（会包含 china.js）
map_info = get_chart_dependencies(map_chart)
print(f"\n地图依赖: {map_info['dependencies']}")
print(f"地图 URLs: {map_info['cdn_urls']}")

# 显示地图（自动加载 china.js）
display_chart(map_chart, height="600px")

```

```python

# ========== 使用不同 CDN ==========
from pyecharts_iframe_renderer import render_chart

# 使用 jsDelivr CDN
render_chart(bar, cdn_provider="jsdelivr")

# 使用自定义 CDN
custom_cdn = "https://my-company-cdn.com/echarts/"
render_chart(bar, custom_cdn_base=custom_cdn)


# ========== 批量渲染多个图表（依赖智能去重） ==========
from pyecharts_iframe_renderer import EChartsRenderer

renderer = EChartsRenderer()

# 多个地图图表，china.js 只会加载一次
map1 = Map().add("数据1", [("北京", 100)], "china")
map2 = Map().add("数据2", [("上海", 200)], "china")

renderer.display(map1, height="400px")
renderer.display(map2, height="400px")  # 复用已加载的 china.js


# ========== 调试：查看依赖加载状态 ==========
from pyecharts_iframe_renderer import get_renderer

renderer = get_renderer()

# 获取依赖信息
deps = renderer.get_dependencies_info(bar)
print(f"\n依赖详情:")
print(f"  - 依赖列表: {deps['dependencies']}")
print(f"  - CDN 提供商: {deps['cdn_provider']}")
print(f"  - CDN 基础URL: {deps['cdn_base']}")
print(f"  - HTML标签:")
print(deps['html_tags'])


# ========== 强制重新加载依赖 ==========
from pyecharts_iframe_renderer import render_chart

# force_reload=True 会重新加载 JS（即使之前加载过）
render_chart(bar, force_reload=True)
```
